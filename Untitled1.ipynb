```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
from selenium.webdriver.common.proxy import Proxy, ProxyType, ProxyTypeFactory
```

```python
s = Service("C:/Users/HITESH/Desktop/chromedriver.exe")

driver = webdriver.Chrome(service = s)
```

```python
driver.get("http://google.com")

time.sleep(0.5)
```

```python
# fetch the search input box using xpath
user_input = driver.find_element(by=By.XPATH,
                                 value='/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
user_input.send_keys('naukari.com')
time.sleep(1)

user_input.send_keys(Keys.ENTER)
time.sleep(1)
```

```python
link = driver.find_element(by=By.XPATH,value='//*[@id="rso"]/div[1]/div/div/div/div/div/div/div/div[1]/a/h3')
link.click()
time.sleep(3)
```

```python
naukari_search = driver.find_element(by=By.XPATH, value='//*[@id="root"]/div[7]/div/div/div[1]/div/div/div/input')
naukari_search.send_keys('Machine learning engineer, Data science')
naukari_search.send_keys(Keys.ENTER)
time.sleep(60)
```

```python
driver.find_element(by=By.XPATH,value='//*[@id="root"]/div[4]/div/div/section[2]/div[3]/div/a[2]/span').click()
time.sleep(60)
```

```python
page_counter = 2
while True:
    driver.find_element(by=By.XPATH,value='//*[@id="root"]/div[4]/div/div/section[2]/div[3]/div/a[2]/span').click()
    time.sleep(60)
    page_counter += 1
    
    print(page_counter)
```
