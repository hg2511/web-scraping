```python
final_dataframe = pd.DataFrame()

for j in range(1,334):
    
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage,'lxml')    
    
    company = soup.find_all('div', class_ = "company-content-wrapper")    
    
    company_name = []
    company_rating = []
    company_reviews = []
    company_old = []
    company_hq = []
    company_emp = []
    company_type = []



    for i in company:
        try:
            company_name.append(i.find('h2').text.strip())    
        except:
            company_name.append(np.nan)
        try:    
            company_rating.append(i.find('p', class_ = 'rating').text.strip())
        except:
            comapny_rating.append(np.nan)
#     company_time.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())    
        
        try:
            company_reviews.append(i.find('a', class_ = 'review-count').text.strip())
        except:
            company_reviews.append(np.nan)
        try:                          
            company_type.append(i.find_all('p', class_ = 'infoEntity')[0].text.strip())
        except:
            company_type.append(np.nan)
        try:
            company_hq.append(i.find_all('p', class_ = 'infoEntity')[2].text.strip())    
        except:
            company_hq.append(np.nan)
        try:    
            company_old.append(i.find_all('p', class_ = 'infoEntity')[1].text.strip())
        except:
            company_old.append(np.nan)
        try:
            company_emp.append(i.find_all('p', class_ = 'infoEntity')[3].text.strip())    
        except:
            company_emp.append(np.nan)
    
    d = {'name': company_name, 'rating': company_rating, 'review': company_reviews, 'type':company_type,'HQ': company_hq, 'how_old': company_old, 'employee':company_emp}    
    df = pd.DataFrame(d)    
    final_dataframe = final_dataframe.append(df, ignore_index=True)
```
